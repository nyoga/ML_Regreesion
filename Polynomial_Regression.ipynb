# Polynomial Regression

## Importing the libraries

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Importing the dataset

```
df = pd.read_csv('Position_Salaries.csv')
X = df.iloc[:, 1:-1].values
y = df.iloc[:, -1].values
```

```
type(X)
```

```
print(np.concatenate((X.reshape(len(X),1),y.reshape(len(y),1)),1))
```

## Training the Linear Regression model on the whole dataset

```
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X, y)
```

## Training the Polynomial Regression model on the whole dataset

```
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=4)
X_poly = poly_reg.fit_transform(X)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, y) #here the instance of LinearRegression() will have default as f(IV1)= f(DV)
```

## Visualising the Linear Regression results

```
plt.scatter(X, y, color='green')
plt.plot(X, regressor.predict(X), color='blue')
plt.title('Truth or Biuff (Linear regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

## Visualising the Polynomial Regression results

```
plt.scatter(X, y, color='green')
plt.plot(X, lin_reg_2.predict(X_poly), color='blue')
plt.title('Truth or Biuff (Linear regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

## Visualising the Polynomial Regression results (for higher resolution and smoother curve)

```
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape((len(X_grid), 1))
x_ploy_grid = poly_reg.fit_transform(X_grid)
plt.scatter(X, y, color='red')
plt.plot(X_grid, lin_reg_2.predict(x_ploy_grid), color='blue')
plt.title('Truth or Biuff (Linear regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

## Predicting a new result with Linear Regression

```
regressor.predict([[6.5]])
```

## Predicting a new result with Polynomial Regression

```
lin_reg_2.predict(poly_reg.fit_transform([[6.5]]))
```
